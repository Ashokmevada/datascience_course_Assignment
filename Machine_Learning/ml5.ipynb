# Q1

Data encoding, also known as data transformation or feature encoding, is the process of converting categorical data into a numerical format that can be used by machine learning algorithms. Categorical data consists of non-numeric values representing different categories or labels. Since many machine learning algorithms work with numerical data, encoding categorical data allows us to use these algorithms effectively.

Data encoding is useful in data science for several reasons:
- Machine learning algorithms typically require numerical inputs, so encoding categorical data makes it compatible with these algorithms.
- Encoding helps capture the ordinal or nominal nature of categorical features, enabling algorithms to learn patterns and make better predictions.
- It facilitates easier handling and manipulation of data during various data preprocessing and analysis steps.
- Data encoding allows us to leverage the full power of mathematical and statistical techniques in data analysis and modeling.

# Q2

Nominal encoding is a type of data encoding used for categorical variables where there is no inherent order or hierarchy among the categories. In nominal encoding, each category is assigned a unique integer value, which represents the category in the transformed numerical format.

Example: Suppose we have a dataset of fruit types, and one of the categorical features is "Fruit Type," with categories "Apple," "Banana," "Orange," "Mango," and "Grapes." We can use nominal encoding to assign each category a unique integer value as follows:
- Apple: 0
- Banana: 1
- Orange: 2
- Mango: 3
- Grapes: 4

# Q3 

Nominal encoding is preferred over one-hot encoding in situations where the categorical variable has a large number of unique categories. One-hot encoding creates a binary feature for each category, resulting in a sparse and high-dimensional dataset, which can be computationally expensive and memory-consuming.

Example: Consider a dataset with a "Country" feature that contains the names of different countries. If there are hundreds of unique countries in the dataset, using one-hot encoding would create hundreds of new binary features, leading to a sparse matrix. In such cases, nominal encoding is preferred as it represents the categorical data efficiently with a single numerical feature for each category.
