```python
category_metric = google_df.groupby('Category').agg({
    'Rating' : 'mean',
    'Installs' : 'sum',
    'Reviews' : 'sum'
}).sort_values(by = 'Rating' , ascending=False).reset_index()
```

```python
plt.figure(figsize =(25,8))
ax = sns.barplot(data = category_metric , x = 'Category'  ,y = 'Rating' , label='Average Rating' )
ax.bar(category_metric.index , category_metric['Installs']/1000 , color = 'green' , label = "Total Installs")
ax.bar(category_metric.index, category_metric['Reviews'] / 1000000, color='orange', label='Total Reviews (Millions)')

ax.set_xticklabels(category_metric['Category'], rotation=90)
ax.set_xlabel('App Category')
ax.set_ylabel('Metric')
ax.set_title('Category Performance: Ratings, Installs, and Reviews')

# Add a legend
ax.legend()
plt.xticks(rotation = 40)
plt.show()
```
