# Q2 

There is no missing values in the dataset

Advantage and disadvantage of various imput techniques

**Advantages**

1. Mean/Median Imputation:

    Advantages:

    Simple and quick to implement.
    Works well with small amounts of missing data.

    Disadvantages:

    Ignores relationships between variables, potentially distorting distributions.
    Can lead to biased estimates if data is not missing at random.
2. Mode Imputation:

    Advantages:

    Suitable for categorical data.
    Simple and quick.

    Disadvantages:
    
    Ignores relationships between variables.
    May not be appropriate for continuous data.

```python
X = wine_df.drop('quality' , axis =1 )
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
pca = PCA()
scaler = StandardScaler()

X = wine_df.drop('quality' , axis = 1)

X_scaled = scaler.fit_transform(X)

X_pca = pca.fit_transform(X_scaled)

```

```python
pca.components_
explained_ratio = pca.explained_variance_ratio_
```

```python
pca_componets , explained_variance_ratio = 0,0

for variance_ratio in explained_ratio:
    explained_variance_ratio += variance_ratio
    pca_componets += 1
    if explained_variance_ratio >= 0.9:
        break

print(f"Minimum number of component required to explain 90 % of the variance: {pca_componets}")
```
